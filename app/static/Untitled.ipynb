```python
from PIL import Image, ImageCms

img = Image.open("cp_face.png").convert("RGB")
# Convert to sRGB if a profile is present
try:
    srgb = ImageCms.createProfile("sRGB")
    img = ImageCms.profileToProfile(img, ImageCms.getOpenProfile(img.info.get("icc_profile")), srgb, outputMode="RGB")
except Exception:
    pass  # no/unsupported profile; we already converted to RGB

img.save("cp_face_srgb.png", optimize=True, quality=95)  # use this file in the app
```
